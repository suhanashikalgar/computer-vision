```python
# Install dependencies
!pip install spotipy gradio opencv-python-headless


import cv2
import numpy as np
import gradio as gr  #
import random #random ‚Äî used for placeholder emotion detection (random choice).
import os
import spotipy #Spotify API client.
from spotipy.oauth2 import SpotifyOAuth #class for Spotify OAuth flows (used when you need user-level permissions).

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials #class for client-credentials flow (app-level access, no user login).

# Replace with your actual client credentials
client_id = "5a49208c06b4449abbaa765c9e896d03"
client_secret = "97a8aac732c7450e8e9bde4af134729a"

#Creates an auth_manager object using the Client Credentials flow. This flow gives application-level access to search and read public data,
# but cannot access user-specific endpoints (playlists, user top tracks, etc.).

# Authenticate using client credentials (no user login)
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager) # creates a spotipy.Spotify client instance you use to call the Spotify API (e.g., sp.search(...))


# --------------------------------------
# üß† EMOTION DETECTION (simplified placeholder)
# --------------------------------------
# You can replace this with your own model (FER, DeepFace, etc.)
def detect_emotion(image):
    # For demonstration, randomly choose an emotion
    emotions = ["happy", "sad", "angry", "surprised", "neutral", "fear", "disgust"]
    detected = random.choice(emotions)
    return detected

# --------------------------------------
# üéµ GET MUSIC FROM SPOTIFY BY EMOTION
# --------------------------------------
#Currently it does not analyze the image ‚Äî it simply selects a random emotion from
def get_music_by_emotion(emotion):
    try:
        # Emotion-to-genre mapping to ensure Spotify results exist
        emotion_map = {
            "happy": "pop",
            "sad": "acoustic",
            "angry": "rock",
            "surprised": "electronic",
            "neutral": "chill",
            "fear": "ambient",
            "disgust": "metal"
        }
        query = f"{emotion_map.get(emotion.lower(), 'pop')} music"

        results = sp.search(q=query, type='track', limit=5)

        if not results['tracks']['items']:
            return f"No Spotify results found for emotion '{emotion}'."

        # Format as a list of clickable links
        recommendations = []
        for track in results['tracks']['items']:
            name = track['name']
            artist = track['artists'][0]['name']
            url = track['external_urls']['spotify']
            recommendations.append(f"{name} - {artist}\nüîó [Listen on Spotify]({url})")

        return "\n\n".join(recommendations)

    except Exception as e:
        return f"Error fetching music: {e}"

# --------------------------------------
# üé§ COMBINE EMOTION DETECTION + RECOMMENDATION
# --------------------------------------
def recommend_music(image):
    if image is None:
        return "Please upload or capture an image."

    emotion = detect_emotion(image)
    music_recs = get_music_by_emotion(emotion)
    return f"üß† Detected Emotion: **{emotion.capitalize()}**\n\nüéµ Recommended Tracks:\n\n{music_recs}"

# --------------------------------------
# üé® GRADIO UI
# --------------------------------------
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("## üéß Emotion-Based Music Recommendation System")
    gr.Markdown("Upload a face image or use your webcam ‚Äî we'll detect your emotion and suggest matching Spotify tracks.")

    with gr.Row():
        input_image = gr.Image(label="Upload or Capture Image", type="numpy")
        output_text = gr.Markdown(label="Recommendations")

    run_btn = gr.Button("üéµ Recommend Music")
    run_btn.click(fn=recommend_music, inputs=input_image, outputs=output_text)

# Launch app
app.launch()
```
